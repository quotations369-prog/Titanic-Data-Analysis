```python
# Titanic Data Analysis - Colab Ready

# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset directly from GitHub raw link
url = "https://raw.githubusercontent.com/quotations369-prog/Titanic-Data-Analysis.../refs/heads/main/titanic.csv"
data = pd.read_csv(url)

# Step 3: Preview dataset
print("First 5 rows of the dataset:")
data.head()

# Step 4: Check for missing values
print("\nMissing values in each column:")
print(data.isnull().sum())

# Step 5: Fill missing Age values with mean
data['Age'].fillna(data['Age'].mean(), inplace=True)

# Step 6: Basic statistics
print("\nBasic statistics:")
print(data.describe())

# Step 7: Survival counts
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=data)
plt.title("Total Survived vs Not Survived")
plt.show()

# Step 8: Survival by Gender
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title("Survival by Gender")
plt.show()

# Step 9: Survival by Passenger Class
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', hue='Pclass', data=data)
plt.title("Survival by Passenger Class")
plt.show()

# Step 10: Average Age by Survival
plt.figure(figsize=(6,4))
sns.kdeplot(data[data['Survived']==1]['Age'], label='Survived', shade=True)
sns.kdeplot(data[data['Survived']==0]['Age'], label='Not Survived', shade=True)
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Density")
plt.legend()
plt.show()

# Step 11: Key Insights
print("\nKey Insights:")
print("- Women had a higher survival rate than men.")
print("- 1st class passengers survived more than 3rd class.")
print("- Children had slightly better chances than adults.")
```
